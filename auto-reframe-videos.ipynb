Add Github Repo Info

```python
import os

github_repo_url = "https://github.com/minhung1126/auto-reframe-videos"
drive_repo_path = "/content/drive/MyDrive/minhung.photography/"
repo_name =  github_repo_url.split('/')[-1].replace('.git', '')
local_repo_path = os.path.join(drive_repo_path, repo_name)

print(f"GitHub 專案網址: {github_repo_url}")
print(f"Google Drive 本地路徑: {local_repo_path}")
```

Git Clone

```python
if not os.path.exists(local_repo_path):
  print(f"專案不存在於 '{local_repo_path}'，將執行'git clone'...")
      # 如果不存在，就建立上層目錄並 clone 整個專案
  os.makedirs(drive_repo_path, exist_ok=True)
  !git clone {github_repo_url} {local_repo_path}
else:
  print(f"專案已存在，將進入目錄並執行 'git pull'...")
  # 如果已存在，就進入該目錄並拉取最新的變更
  %cd {local_repo_path}
  !git pull

  print("\n✅ 同步完成！")
```
